```python
#Task 2 - The cost column in the dataframe “Project” has some missing values. 
        #Computing these missing values and replacing with running avergae using "for" loop.

project_df = pd.read_csv(project_path)

project_df
```

```python
total_sum = 0
count = 0

for i in range(len(project_df)):
    
    if pd.isna(project_df.loc[i, "Cost"]):
        
        average = total_sum/count
        
        project_df.loc[i, "Cost"] = average
    
    else:
        total_sum += project_df.loc[i, "Cost"]
        
        count += 1
```

```python
#Task 4 - Joining all three dataframes in one single dataframe and naming it "final"

employee_df = pd.read_csv(employee_path)

seniority_df = pd.read_csv(seniority_path)

project_df = pd.read_csv(project_path)

employee_df, seniority_df, project_df
```

```python
emp_seniority_df = pd.merge(employee_df, seniority_df, on="ID", how="inner")

emp_seniority_df
```

```python
final_df = pd.merge(emp_seniority_df, project_df, on="ID", how="inner")

final_df
```

```python
#Task 5 - Adding a new bonus column in the Final dataframe. Giving a 5% bonus concerning project cost to the employees who finished projects

final_df
```

```python
final_df["Bonus"] = np.where(
    final_df["Status"] == "Finished",
    final_df["Cost"] * 0.05,
    0
)
```

```python
final_df
```

```python
final_df.to_csv(os.path.join(base_path,"Final.csv"), index=False)
```

```python
#Task 6 - Demoting the designation of those with project status "fail" by 1 and deleting records of those with designation level above 4.

final_df = pd.read_csv("Final.csv")

final_df
```

```python
for i in range(len(final_df)):
    if final_df.loc[i, "Status"] == "Failed":
        final_df.loc[i, "Designation Level"] = final_df.loc[i, "Designation Level"] + 1
    else:
        pass
```

```python
valid_rows = []

for i in range(len(final_df)):
    if final_df.loc[i, "Designation Level"] <= 4:
        valid_rows.append(i)
    else:
        pass
```

```python
final_df = final_df.loc[valid_rows]

final_df.reset_index(drop=True, inplace=True)
```

```python
#Task 7 - Adding “Mr.” and “Mrs.” to the first name column and dropping the gender column.

for i in range(len(final_df)):
    if final_df.loc[i, "Gender"] == "M":
        final_df.loc[i, "First Name"] = "Mr. " + final_df.loc[i, "First Name"]
    elif final_df.loc[i, "Gender"] == "F":
        final_df.loc[i, "First Name"] = "Mrs. " + final_df.loc[i, "First Name"]
```

```python
final_df.drop("Gender", axis=1, inplace=True)
```

```python
final_df
```

```python
final_df.to_csv("Final.csv", index=False)
```

```python
#Task 8 - Promoting designation level by 1 for the employees whose age is more than 29 years using IF condition.

final_df.columns
```

```python
for i in range(len(final_df)):
    
    if final_df.loc[i, "Age"] > 29:
        final_df.loc[i, "Designation Level"] = final_df.loc[i, "Designation Level"] - 1
    else:
        pass
```

```python
final_df
```

```python
final_df.to_csv("Final.csv", index=False)
```

```python
#Task 9 - Adding the cost of all projects for each Employee and saving it in new dataframe “TotalProjCost” with three columns ID, First Name, and Total cost


total_cost_df = final_df.groupby("ID")["Cost"].sum().reset_index()
```

```python
name_df = final_df[["ID", "First Name"]].drop_duplicates()
```

```python
TotalProjCost = pd.merge(name_df, total_cost_df, on="ID", how="inner")
```

```python
TotalProjCost.rename(columns={"Cost": "Total Cost"}, inplace=True)
```

```python
TotalProjCost
```

```python
TotalProjCost.to_csv("TotalProjCost.csv", index=False)
```
